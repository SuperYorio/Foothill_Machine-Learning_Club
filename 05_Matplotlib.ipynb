```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.style.use('classic')
```

```python
import numpy as np
```

```python
# Avoid using multiple show()'s
# plt.show() is generally used in scripts instead of Notebooks
x = np.linspace(0, 100, 10)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

```python
# Using the seaborn style
plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
# Different patterns for your graph
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
# Setting limits on your graph
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
# Specify the axis using the .axis() method
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

```python
# Setting labels
plt.plot(x, np.sin(x))
plt.title("Sine Curve")
plt.xlabel("x")
plt.ylabel("y = sin(x)");
```

```python
# To show the legend
plt.plot(x, np.sin(x), '-r', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend();
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, 'o', color='red');
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="Marker='{0}'".format(marker))
    plt.legend(numpoints=1)
    plt.xlim(0, 1.8);
```

```python
# Better yet, just use the scatter() method
plt.scatter(x, y, marker='d')
```
